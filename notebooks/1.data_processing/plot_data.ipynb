### Data Distribution
This notebook consisted of code where we could plot the data distribution for each class from each partner. Used this for the plotting of entire dataset, training datset and validation dataset.

```python
# creating the dataset
values_1 = [216, 244, 40, 50, 1, 107, 12, 208]
data = {'Bayer': values_1[0], 'Merck': values_1[1], 'AstraZeneca': values_1[2], 'Ksilink': values_1[3], 'Servier': values_1[4], 'MPI': values_1[5], 'Pfizer': values_1[6], 'Broad Institute': values_1[7]}
courses = list(data.keys())
values = list(data.values())
  
fig = plt.figure(figsize = (12, 6))
 
# creating the bar plot
bars = plt.bar(courses, values, color ='maroon', width = 0.4)
 
index = 0
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x()+0.08, yval+2, values_1[index])
    index += 1
    
plt.xlabel("Partners")
plt.ylabel("Number of images")
plt.title("Data Distribution for Debris class")
plt.savefig("data_debris.png")
plt.show()
```
